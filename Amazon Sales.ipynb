# Explolatory Data Analysis

```python
ax=sns.countplot(x="Size",data=df)
plt.show()
```

```python
ax=sns.countplot(x="Size",data=df)
#for bars in lt.containers:
 #   lt.bar_label(bars)
ax.bar_label(ax.containers[0])    
```

```python
gb=df.groupby("Status").agg({"Amount":"mean","ship-city":"count"})
#print(gb)
sns.heatmap(data=gb,annot=True)
plt.show()
```

```python
ax=sns.countplot(data=df,x="Category")
ax.bar_label(ax.containers[0])
plt.xticks(rotation=32)
plt.show()
```
